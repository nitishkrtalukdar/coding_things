```python
##OOPS

##Class is a blueprint for creating objects. It defines a set of attributes and methods that the objects created from the class will have. An object is an instance of a class, which means it is a specific realization of the class with its own unique data.

##Example of a class and an object in Python:

class Car:
    def __init__(self,brand,speed):
        self.brand=brand
        self.speed=speed
    def accelerate(self):
        self.speed+=10
        return f"{self.brand} going at {self.speed} km/h"

my_car=Car("Ferrari",200)
your_car=Car("lambo",120)

print(my_car.accelerate())  # Output: Ferrari going at 210 km/h
print(your_car.accelerate())  # Output: lambo going at 130 km/h 
```
